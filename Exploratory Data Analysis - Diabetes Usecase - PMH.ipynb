```python
dbt_data_copy=dbt_data.copy()
```

```python
dbt_data_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']]=dbt_data_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.NaN)
```

```python
dbt_data_copy
```

```python
dbt_data_copy.isnull().sum()
```

```python
pl=dbt_data_copy.hist(figsize=(12,12))
```

```python
dbt_data_copy['Glucose'].fillna(dbt_data_copy['Glucose'].mean(),inplace=True)
```

```python
dbt_data_copy.info()
```

```python
dbt_data_copy.isnull().sum()
```

```python
dbt_data_copy['BloodPressure'].fillna(dbt_data_copy['BloodPressure'].mean(),inplace=True)
```

```python
dbt_data_copy.isnull().sum()
```

```python
dbt_data_copy['SkinThickness'].fillna(dbt_data_copy['SkinThickness'].mean(),inplace=True)
```

```python
dbt_data_copy.isnull().sum()
```

```python
dbt_data_copy['Insulin'].fillna(dbt_data_copy['Insulin'].mean(),inplace=True)
```

```python
dbt_data_copy.isnull().sum()
```

```python
dbt_data_copy['BMI'].fillna(dbt_data_copy['BMI'].mean(),inplace=True)
```
